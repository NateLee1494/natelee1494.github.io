````python
import numpy as np
np.sort(df["Year Acquired"].unique())
````

````python
brush = alt.selection_interval(encodings = ['x'])

df_clean = df[(df["Year Acquired"] >= 1753)]

upper = alt.Chart(df_clean).transform_filter(brush).mark_bar().encode(
    x = "Agency Name:N",
    y = "Square Footage:Q",
    color = "Agency Name:N"
)

lower = alt.Chart(df_clean).transform_filter({
    "field" : "Year Acquired", "range" : [1753, 2019]}).mark_point().encode(
    x = alt.X('Year Acquired:Q', scale = alt.Scale(domain = [1753, 2019])),
    y = 'Square Footage:Q'
).add_params(brush)

chart2 = upper & lower

chart2
````

````python
chart2.save("chart2.json")
chart2.save("chart2.png")
````
